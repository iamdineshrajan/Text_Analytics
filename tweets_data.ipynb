```python
from nltk.tokenize import word_tokenize
```

```python
token=word_tokenize(t2)
```

```python
from nltk.corpus import stopwords
```

```python
t3=set(stopwords.words('english'))
```

```python
t4=[word for word in token if word not in t3]
```

```python
t4
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
t5=WordNetLemmatizer()
```

```python
t6=[]
```

```python
for word in t4:
    t6.append(t5.lemmatize(word))
    print(t5.lemmatize(word))
```

```python
from nltk.probability import FreqDist
```

```python
t7=FreqDist(t6)
```

```python
t7.tabulate()
t7
```
